###  Visualize Key Insights

```python
color = [
    "#FF5733",  # Red-Orange
    "#33FF57",  # Lime Green
    "#3357FF",  # Blue
    "#F1C40F",  # Yellow
    "#8E44AD",  # Purple
    "#E67E22",  # Orange
    "#1ABC9C",  # Teal
    "#2ECC71",  # Green
    "#3498DB",  # Sky Blue
    "#E74C3C",  # Red
]
```

```python
#Count of Movies vs TV Shows

plt.figure(figsize=(3,4))
ax = sns.countplot(data=df, x='type', hue='type', palette='Set1', legend=False)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2, height),
                ha='center')
plt.title('Count of Movies vs TV Shows')
plt.xlabel('Program Types')
plt.tight_layout()
plt.show()
```

```python
#Top 10 Countries with Most Content

plt.figure(figsize=(9,5))
top_countries = df['country'].value_counts().head(10)
bars = top_countries.plot(kind='barh', color=color, edgecolor='#34495E')
for index, value in enumerate(top_countries):
    plt.text(value + 2, index, str(value), va='center', fontsize=9)
plt.title('Top 10 Countries with Highest Number of Shows')
plt.xlabel('Number of Shows')
plt.ylabel("Country's Name")
plt.tight_layout()
plt.show()
```

```python
# Content Added Over the Years

plt.figure(figsize=(9,4))
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('Number of Titles Added per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
# Most Common Genres

genres = df['listed_in'].dropna().apply(lambda x: x.split(', '))
all_genres = Counter([genre for sublist in genres for genre in sublist])
top_genres = pd.DataFrame(all_genres.most_common(10), columns=['Genre', 'Count'])

sns.barplot(data=top_genres, x='Count', y='Genre',hue='Count', palette=color)
plt.title('Top 10 Netflix Genres')
plt.show()
```

```python
# Comparison Movies between TV Shows

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
movie_counts = df[df['type'] == 'Movie']['release_year'].value_counts()
ax[0].scatter(movie_counts.index, movie_counts.values, color='blue', s=20, marker='*')
ax[0].set_title('Release Year vs Number of Movies')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('No. of Movies')

tv_shows_counts = df[df['type'] == 'TV Show']['release_year'].value_counts()
ax[1].scatter(tv_shows_counts.index, tv_shows_counts.values, color='orange', s=20, marker='+')
ax[1].set_title('Release Year vs Number of TV Shows')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('No. of TV Shows')

fig.suptitle('Comparison Between Movies and TV Shows', fontsize=20)
plt.tight_layout() 
plt.show()
```

```python
# Number of Release changed over the year

plt.figure(figsize=(10,4))
value_counts = df['release_year'].value_counts()
plt.plot(value_counts.index, value_counts.values,color='blue',linestyle=':',linewidth=2,marker='*',markerfacecolor='red',markeredgecolor='red')
plt.title('Number of Release changed over the year')
plt.xlabel('Year')
plt.ylabel('No. of Movies')
plt.tight_layout()
plt.show()
```
